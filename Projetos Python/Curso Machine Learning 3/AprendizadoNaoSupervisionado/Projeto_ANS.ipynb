```python
import pandas as pd
```

```python
dados = pd.read_csv('dados-padaria.csv')
```

```python
dados.info
```

```python
dados.head()
```

```python
transacao = []
for item in dados ['TransactionNo'].unique():
    lista = list(set(dados[dados['TransactionNo'] == item]['Items']))
    transacao.append(lista)
```

```python
transacao[0:5]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
transacao_te = te.fit(transacao).transform(transacao)
```

```python
transacao_te
```

```python
transacao_transformado = pd.DataFrame(transacao_te, columns = te.columns_)
transacao_transformado
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
itens_frequentes_apriori = apriori(transacao_transformado, use_colnames = True, min_support=0.02)
itens_frequentes_apriori.sort_values(['support'], ascending = False)
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
regras_apriori = association_rules(itens_frequentes_apriori, metric = 'confidence', min_threshold = 0.5)
regras_apriori
```
